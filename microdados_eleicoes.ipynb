```python
# consultando totais
total_ver_2004 = len(ver_2004.index)
total_ver_2008 = len(ver_2008.index)
total_ver_2012 = len(ver_2012.index)
total_ver_2016 = len(ver_2016.index)

total_pft_2004 = len(pft_2004.index)
total_pft_2008 = len(pft_2008.index)
total_pft_2012 = len(pft_2012.index)
total_pft_2016 = len(pft_2016.index)

total_sen_2006 = len(sen_2006.index)
total_sen_2010 = len(sen_2010.index)
total_sen_2014 = len(sen_2014.index)
total_sen_2018 = len(sen_2018.index)

total_gov_2006 = len(gov_2006.index)
total_gov_2010 = len(gov_2010.index)
total_gov_2014 = len(gov_2014.index)
total_gov_2018 = len(gov_2018.index)
```

```python
anos_1 = ['2004', '2008', '2012', '2016']
anos_2 = ['2006', '2010', '2014', '2018']

x = np.arange(len(anos_1))  # posicao das labes
x2 = np.arange(len(anos_2))  # posicao das labes
width = 0.35  # width das barras

## legenda de quantidade a cima das barras
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```
